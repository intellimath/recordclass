```python
from recordclass import dataobject
```

```python
class extended_dataobject(dataobject):
    def __init__(self, *args, **kwargs): 
        super().__init__(*args, **kwargs)
        
class Param(extended_dataobject):
    x:dict = {'a':1}
    y:tuple = (0,0)
    
    # def __init__(self, x=-1, y=-2):
    #     self.x = x
    #     self.y = y        
```

```python
print(Param.__init__)
print(Param.__new__)
```

```python
p = Param.__new__(Param, 1, 2)
print(p)
p.__init__(1, 2)
print(p)
```

```python
p = Param(x=1, y=2)
print(p)
```

```python
from pickle import dumps, loads, Pickler, Unpickler
```

```python
print(p)
print(p.__reduce__())
s = dumps(p)
p1 = loads(s)
print(type(p1.x))
print(p1)
```

```python
import io
sfile = io.StringIO()
Pickler(sfile).dump(p)
s = sfile.getvalue()
print(s)
# p1 = Unpickler()
```
