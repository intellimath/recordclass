```python
from recordclass import dataobject
```

```python
class HGeo(dataobject):
    latitude:double
    longitude:double
        
    def name(self):
        return "{0}, {1}".format(self.latitude, self.longitude)

class HAddr(dataobject):
    street_address:str
    locality:str
    region:str
    country_name:str
    postal_code:str
    geo:HGeo = None
        
    def name(self):
        values = (getattr(self, fname, None) for fname in self.__fields__)
        return " ".join(v for v in values if v is not None)
    
class HEntry(dataobject):
    entry_title:str
    author:str
    updated:object
    entry_content:str = None
    entry_summary:str = None
    published:object = None
    bookmark:str = None
    tags:list = None
    
```
